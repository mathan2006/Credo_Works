```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import roc_auc_score
from sklearn.preprocessing import Binarizer
from sklearn.metrics import confusion_matrix
from sklearn.metrics import precision_recall_fscore_support
import pandas as pd
import numpy as np
```

```python
X = pd.read_csv("Car_Model.csv")
```

```python
X.head()
```

```python
#Taking y feature
y=X.pop("mpg")
```

```python
X.describe(include='all')
```

```python
#X['Horsepower']=X['Horsepower'].replace('?','NaN')
#X.iloc[25:35]
#X['Horsepower']=X['Horsepower'].astype(float)
X['Horsepower'].fillna((X['Horsepower'].mean()), inplace=True)
```

```python
X.dtypes
#X['Horsepower']=X['Horsepower'].astype(int)
```

```python
X.describe(include='all')
```

```python
X.describe()
```

```python
#Dropping unwanted features
X=X.drop('Carname',1)
```

```python
X.describe(include='all')
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2)
```

```python
#Regressor model fit
model = RandomForestRegressor(n_estimators=1000,random_state=123)
model.fit(X_train,y_train)
y_pred=model.predict(X_test)
```

```python
from sklearn import metrics
print("MAE: ",metrics.mean_absolute_error(y_test,y_pred))
print("MSE: ",metrics.mean_squared_error(y_test,y_pred))
print("RMSE: ",np.sqrt(metrics.mean_squared_error(y_test,y_pred)))
```
