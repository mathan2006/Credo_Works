```python
def lammitize_with_tag(sentence):
    sent=TextBlob(sentence)
    tag_dict={"J":'a',
              "N":'n',
              "V":'v',
              "R":'r'}
    words_tags=[(w, tag_dict.get(pos[0], 'n')) for w, pos in sent.tags]
    lemmatized_words = [wd.lemmatize(tag) for wd, tag in words_tags]
    return " ".join(lemmatized_words)
```

```python
data['text']=data.text.apply(lammitize_with_tag)
```
